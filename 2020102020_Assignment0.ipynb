```python
fig = plt.figure(figsize=(20,20))

sp = fig.add_subplot(221)
plt.imshow(img1)
plt.axis('off') 

sp = fig.add_subplot(222)
plt.imshow(img2)
plt.axis('off')

sp = fig.add_subplot(223)
plt.imshow(img3)
plt.axis('off')

sp = fig.add_subplot(224)
plt.imshow(img4)
plt.axis('off')

plt.show()
```

From the images, we can observe that places of red colour and blue colour are inter changed, i.e., where red should be present, we are seeing blue and where blue should be present, we are seeing red.  
This issue arises the input taken by cv will be in BGR form. So, to fix this we should convert into RGB form before the plotting.

```python
fig = plt.figure(figsize=(20,20))

sp = fig.add_subplot(221)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
plt.axis('off') 

sp = fig.add_subplot(222)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.axis('off')

sp = fig.add_subplot(223)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
plt.imshow(img3)
plt.axis('off')

sp = fig.add_subplot(224)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
plt.imshow(img4)
plt.axis('off')

plt.show()
```
